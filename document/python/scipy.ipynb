```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate 
```

$$f(x) = Ae^{-Bx^2}$$

```python
def f(x,A,B):
    return A*np.exp(-B * x ** 2)
```

```python
print("resultado",  ",  error cometido")
print(integrate.quad(f,0,5,args=(1.0,1.0)))
```

$$y'+ y = 0$$
$$y' = f(y)$$

```python
def f(y,t):
    return -y
```

```python
y0 = 1.0
t = np.linspace(0,3)
```

```python
solve = integrate.odeint(f,y0,t)
```

```python
plt.plot(t,solve)
```

$$y'' + y = 0$$

```python
def g(y,t):
    return np.array([y[1],-y[0]])
```

```python
t = np.linspace(0,10)
y0 = np.array([1.0,0.0])
solve = integrate.odeint(g,y0,t)
print("y","y'")
solve
```

```python
plt.plot(t,solve[:,0])
plt.plot(t,solve[:,1],'--')
```

```python
#ecuaciones algebraicas no linales
from scipy import optimize
```

$$M = E - esin(E)$$
$$F(E) = E - esin(E) - M$$

```python
def F(E,e,M):
    return E - e * np.sin(E)-M
```

```python
solve = optimize.root(F,0.1,args=(0.016,0.1))
solve.x
```
