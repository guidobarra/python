```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

$$ f(x) = e^{-x^2}$$

```python
def f(x):
    ### e elevado al -x^2
    return np.exp(-x ** 2)
```

```python
x = np.linspace(-2,5,num=30)
plt.plot(x,f(x),"red",label="Funcion f(x)")#ojo hace le estoy pasando dos array f(x) devuelve un array porque le pase un array PROBA print(f(x))!!!
plt.xlabel("eje x")#nombres de los ejes
plt.ylabel("$f(x)$")
plt.legend()# para que aparesca "Funcion f(x)"
plt.plot(x,f(x),'ob',label="Funcion f(x) p")#le paso el argumento 'o' para obtener los puntos 
plt.grid(True)#para las cuadriculas del plano
plt.title("$f(x) = e^{-x^2}$")
#para informacion ir a help y clickear matplolib
```

```python
fig, axes = plt.subplots()
axes.plot(x,f(x),"or",label = "Funcion f(x)")# un objeto tiene la imagen
fig.savefig("gradica.png")#guardo la imagen en un formato
axes.set_xlim(-4,3)
axes.set_ylim(-2,2)
axes.grid(True)
```

```python
fig, axes = plt.subplots(1,2,sharey=True)#1x2 como una matriz con imagenes ponele, sharey es para que compartan el eje y
axes[0].plot(x,f(x),"blue",label="Funcion f(x)")
axes[1].plot(x,-f(x),"red",label="Funcion -f(x)")

axes[0].set_xlabel("eje x blue")
axes[1].set_xlabel("eje x red")
axes[0].grid(True)
axes[1].grid(True)
```

$$g(x,y) = \cos(x) + \sin(y)^2$$

```python
x = np.linspace(-2,2)
y = np.linspace(-2,2)
xx,yy = np.meshgrid(x,y)
```

```python
def g(x,y):
    return np.cos(x)+np.sin(y) ** 2
```

```python
zz = g(xx,yy)
```

```python
fig, axes = plt.subplots()
axes.contour(xx,yy,zz)
axes.contour(xx,yy,zz,[-2,0,2])
axes.contour(xx,yy,zz,np.linspace(-1,1),cmap=plt.cm.autumn)
axes.contourf(xx,yy,zz,np.linspace(-1,1),cmap=plt.cm.autumn)
```
